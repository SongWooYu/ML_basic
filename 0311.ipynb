```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as ss
 
b = ss.distributions.binom
 
for flips in [5, 10, 20, 40, 80]:
    successes = np.arange(flips)
    out_distribution = b.pmf(successes, flips, 0.5)
    plt.hist(successes, flips, weights=out_distribution)
   
plt.xlim(0, 55)
 
plt.show()
```

```python
quantity = [2, 12, 3]
costs = [12.5, 0.5, 1.75]
partical_cost = []
for q, c in zip(quantity, costs):
    partical_cost.append(q * c)
print(sum(partical_cost))
```

```python
import numpy as np
quantity = [2, 12, 3]
costs = [12.5, 0.5, 1.75]
print(np.array(quantity) * np.array(costs))
print(sum(np.array(quantity) * np.array(costs)))
```

```python
import numpy as np
quantity = np.array([2, 12, 3])
costs = np.array([12.5, 0.5, 1.75])
print(quantity.dot(costs))
print(np.dot(quantity, costs))
print(quantity @ costs)
```

```python
import numpy as np
values = np.array([10.0, 20.0, 30.0])
weights = np.full_like(values, 1/3)
print("weights:", weights)
print("via mean:", np.mean(values))
print("via weights and dot:", np.dot(weights, values))
```

```python
import numpy as np
def is_even(n):
    return n % 2 == 0
winnings = 0.0
for toss_ct in range(10000):
    die_toss = np.random.randint(1, 7)
    winnings += 1.0 if is_even(die_toss) else -0.5
print(winnings)
```
